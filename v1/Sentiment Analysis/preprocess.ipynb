```python
df = pd.DataFrame(columns =['Chapter Name', 'Username', 'Date', 'Text'])
for line in open("cog editedchap.csv", mode='r', encoding='utf-8'):
    comma1 = line.index(',')
    comma2 = line.index(',', comma1+1)
    comma3 = line.index(',', comma2+1)
    chaptername = line[0:comma1]
    username = line[comma1+1:comma2]
    date = line[comma2+1:comma3]
    remainingline = line[comma3+1:]
    stripreply = re.sub("Reply|1 Reply|\d+ Replies", '', remainingline)
    stripcomma = re.sub(",", ' ', stripreply)
    cleantext = ftfy.fix_text(stripcomma)
    mycomment = pd.Series(list([chaptername, username, date, cleantext]), index=df.columns)
    df = df.append(mycomment, ignore_index=True)    
f.closed
```
